<p style="font-size:20px">Import libraries and plot setup.</p>

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from math import pi,sin
```

```python
def getPlotReady(title='',xLabel='', yLabel='', figSize=None):
    fig = plt.figure()
    if figSize != None:
        fig.set_size_inches(figSize[0], figSize[1])
        
    ax = fig.add_subplot(1,1,1)
    ax.set_title(title,fontdict={'fontsize': 20}, pad=15.0)
    ax.set_xlabel(xLabel,fontdict={'fontsize': 16})
    ax.set_ylabel(yLabel,fontdict={'fontsize': 16})
```

```python
def getReadySubplot(numOfPlot, signals, figSize=None):
    fig, axes = plt.subplots(numOfPlot, 1) 
    if figSize != None:
        fig.set_size_inches(figSize[0], figSize[1])
    fig.suptitle('Multiple Signal in subplot.', fontsize=20, y=0.93)
    for index in range(numOfPlot):
        axes[index].plot(signals[index]['x'],signals[index]['y'])
        axes[index].set_title(signals[index]['title'])
        axes[index].set_xlabel(signals[index]['xlabel'])
        axes[index].set_ylabel(signals[index]['ylabel'])
```

<p style="font-size:20px;">$$\begin{aligned}&(i) Unit\ Step\ Signal\end{aligned} $$</p>

```python
def unit_step(times):
    return [1 if t >= 0 else 0 for t in times]
```

```python
times = np.arange(-5,5,0.01)
amp = unit_step(times)
getPlotReady(title='$Unit\ Step\ Signal$', xLabel='Time(s)', yLabel='Amplitude(m)', figSize=(15,8))
plt.plot(times, amp)
```

<p style="font-size:20px;">$$\begin{aligned}&(ii) Unit\ Impuls\ Signal\end{aligned} $$</p>

```python
def unit_impulse(times, position = 0):
    return [1 if abs(t-position)<0.001 else 0 for t in times]
    
```

```python
times = np.arange(-5,5,0.01)
amp = unit_impulse(times, position = 0)
# print(times)
# print(amp)
getPlotReady(title='$Unit\ Impulse\ Signal$', xLabel='Time(s)', yLabel='Amplitude(m)', figSize=(15,8))
plt.plot(times,amp)
```

<p style="font-size:20px;">$$\begin{aligned}&(iii) Ramp\ Signal\end{aligned} $$</p>

```python
def ramp(times):
    return [t if t>=0 else 0 for t in times]
```

```python
times = np.arange(-5,5,0.01)
amp = ramp(times)
getPlotReady(title='$Ramp\ Signal$', xLabel='Time(s)', yLabel='Amplitude(m)', figSize=(15,8))
plt.plot(times,amp)
```

<p style="font-size:20px;">$$\begin{aligned}&(iv) Periodic\ Sinusoidal\ Signal\end{aligned} $$</p>

```python
def sinusoidal_signal(times, omega):
    return [sin(omega * t) for t in times]
```

```python
T = 2
omega = (2 * pi)/T
times = np.arange(0,3*T,0.01)
amp = sinusoidal_signal(times, omega)
getPlotReady(title='$Periodic\ Sinusoidal\ Signal$', xLabel='Time(s)', yLabel='Amplitude(m)', figSize=(15,8))
plt.plot(times,amp)
```

<p style="font-size:20px;">Ploting all the signal in a single figure. </p>

```python
signals = [{'title': '$Unit\ Step\ Signal$', 'x': np.arange(-5,5,0.01),
             'y': unit_step(np.arange(-5,5,0.01)), 'xlabel':'Time(s)', 'ylabel':'Amplitude(m)'},
            {'title': '$Unit\ Impulse\ Signal$', 'x': np.arange(-5,5,0.01),
             'y': unit_impulse(np.arange(-5,5,0.01)), 'xlabel':'Time(s)', 'ylabel':'Amplitude(m)'},
            {'title': '$Ramp\ Signal$', 'x': np.arange(-5,5,0.01),
             'y': ramp(np.arange(-5,5,0.01)), 'xlabel':'Time(s)', 'ylabel':'Amplitude(m)'},
           {'title': '$Periodic\ Sinusoidal\ Signal$', 'x': np.arange(0,3*T,0.01),
             'y': sinusoidal_signal(np.arange(0,3*T,0.01), omega), 'xlabel':'Time(s)', 'ylabel':'Amplitude(m)'}]

getReadySubplot(numOfPlot=4, signals=signals, figSize=(15,20))
plt.show()
```
