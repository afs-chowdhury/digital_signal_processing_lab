<p style="font-size:20px">Import libraries and plot setup.</p>

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from math import pi,sin
```

```python
def getPlotReady(title='',xLabel='', yLabel='', figSize=None):
    fig = plt.figure()
    if figSize != None:
        fig.set_size_inches(figSize[0], figSize[1])
        
    ax = fig.add_subplot(1,1,1)
    ax.set_title(title,fontdict={'fontsize': 20}, pad=15.0)
    ax.set_xlabel(xLabel,fontdict={'fontsize': 16})
    ax.set_ylabel(yLabel,fontdict={'fontsize': 16})
```

```python
def getReadySubplot(numOfPlot, signals, figSize=None):
    fig, axes = plt.subplots(numOfPlot, 1) 
    if figSize != None:
        fig.set_size_inches(figSize[0], figSize[1])
    fig.suptitle('Multiple Signal in subplot.', fontsize=20, y=0.93)
    for index in range(numOfPlot):
        axes[index].stem(signals[index]['x'],signals[index]['y'])
        axes[index].set_title(signals[index]['title'])
        axes[index].set_xlabel(signals[index]['xlabel'])
        axes[index].set_ylabel(signals[index]['ylabel'])
```

<p style="font-size:20px;">$$\begin{aligned}&(i) Unit\ Step\ Signal\end{aligned} $$</p>

```python
def unit_step(times):
    return [1 if times[i] >= 0 else 0 for i in range(len(times))]
```

```python
times = np.arange(-5,20,1)
amp = unit_step(times)
getPlotReady(title='Unit Step Signal', xLabel='Times(s)',yLabel='Amplitude', figSize=(15,8))
plt.stem(times,amp)
```

<p style="font-size:20px;">$$\begin{aligned}&(ii) Unit\ Impuls\ Signal\end{aligned} $$</p>

```python
def unit_impulse(times):
    return [1 if times[i] == 0 else 0 for i in range(len(times))]
```

```python
times = np.arange(-5,20,1)
amp = unit_impulse(times)
getPlotReady(title='Unit Impulse Signal', xLabel='Times(s)',yLabel='Amplitude', figSize=(15,8))
plt.stem(times,amp)
```

<p style="font-size:20px;">$$\begin{aligned}&(iii) Ramp\ Signal\end{aligned} $$</p>

```python
def ramp(times):
    return [times[i] if times[i]>=0 else 0 for i in range(len(times))]
```

```python
times = np.arange(-5,20,1)
amp = ramp(times)
getPlotReady(title='Ramp Signal', xLabel='Times(s)',yLabel='Amplitude', figSize=(15,8))
plt.stem(times,amp)
```
