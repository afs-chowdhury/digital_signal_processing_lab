<p style="font-size:20px">Import libraries and plot setup.</p>

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from math import pi,sin,cos
```

```python
def getPlotReady(title='',xLabel='', yLabel='', figSize=None):
    fig = plt.figure()
    if figSize != None:
        fig.set_size_inches(figSize[0], figSize[1])
        
    ax = fig.add_subplot(1,1,1)
    ax.set_title(title,fontdict={'fontsize': 20}, pad=15.0)
    ax.set_xlabel(xLabel,fontdict={'fontsize': 16})
    ax.set_ylabel(yLabel,fontdict={'fontsize': 16})
```

```python
def getReadySubplot(signals,title = 'Multiple Signal in subplot.', figSize=None):
    numOfPlot = len(signals)
    fig, axes = plt.subplots(numOfPlot, 1) 
    if figSize != None:
        fig.set_size_inches(figSize[0], figSize[1])
    fig.suptitle(title, fontsize=20, y=0.93)
    for index in range(numOfPlot):
        axes[index].plot(signals[index]['x'],signals[index]['y'])
        axes[index].set_title(signals[index]['title'])
        axes[index].set_xlabel(signals[index]['xlabel'])
        axes[index].set_ylabel(signals[index]['ylabel'])
```

<p style="font-size:20px;">$$\begin{aligned}&(i) Addition\end{aligned} $$</p>

```python
def addition(times,omega1,omega2):
    return [cos(omega1 * t) + sin(omega2 * t) for t in times]

def signal1(times,omega):
    return [cos(omega * t) for t in times]

def signal2(times,omega):
    return [sin(omega * t) for t in times]
    
```

```python
T = 4
omega1 = (2 * pi)/T
omega2 = (2 * 2 * pi)/T
times = np.arange(0,3 * T, 0.01)

signals = [{'title':'$cos(t)$','x':times, 'y':signal1(times,omega1), 'xlabel':'Time(s)', 'ylabel':'Amplitude(m)'}, 
           {'title':'$sin(2t)$','x':times, 'y':signal2(times,omega2), 'xlabel':'Time(s)', 'ylabel':'Amplitude(m)'},
           {'title':'$cos(t) + sin(2t)$','x':times, 'y':addition(times,omega1,omega2), 'xlabel':'Time(s)', 'ylabel':'Amplitude(m)'}]

getReadySubplot(title='Addition of Two Signal', signals=signals, figSize=(15,18))
plt.show()
```

<p style="font-size:20px;">$$\begin{aligned}&(ii) Subtraction\end{aligned} $$</p>

```python
def subtraction(times,omega1,omega2):
    return [cos(omega1 * t) - sin(omega2 * t) for t in times]

def signal1(times,omega):
    return [cos(omega * t) for t in times]

def signal2(times,omega):
    return [sin(omega * t) for t in times]
    
```

```python
T = 4
omega1 = (2 * pi)/T
omega2 = (2 * 2 * pi)/T
times = np.arange(0,3 * T, 0.01)

signals = [{'title':'$cos(t)$','x':times, 'y':signal1(times,omega1), 'xlabel':'Time(s)', 'ylabel':'Amplitude(m)'}, 
           {'title':'$sin(2t)$','x':times, 'y':signal2(times,omega2), 'xlabel':'Time(s)', 'ylabel':'Amplitude(m)'},
           {'title':'$cos(t) - sin(2t)$','x':times, 'y':subtraction(times,omega1,omega2), 'xlabel':'Time(s)', 'ylabel':'Amplitude(m)'}]

getReadySubplot(title='Subtraction of Two Signal', signals=signals, figSize=(15,18))
plt.show()
```

<p style="font-size:20px;">$$\begin{aligned}&(iii) Multiplication\end{aligned} $$</p>

```python
def multiplication(times,omega1,omega2):
    return [cos(omega1 * t) * sin(omega2 * t) for t in times]

def signal1(times,omega):
    return [cos(omega * t) for t in times]

def signal2(times,omega):
    return [sin(omega * t) for t in times]
    
```

```python
T = 4
omega1 = (2 * pi)/T
omega2 = (2 * 2 * pi)/T
times = np.arange(0,3 * T, 0.01)

signals = [{'title':'$cos(t)$','x':times, 'y':signal1(times,omega1), 'xlabel':'Time(s)', 'ylabel':'Amplitude(m)'}, 
           {'title':'$sin(2t)$','x':times, 'y':signal2(times,omega2), 'xlabel':'Time(s)', 'ylabel':'Amplitude(m)'},
           {'title':'$cos(t) * sin(2t)$','x':times, 'y':multiplication(times,omega1,omega2), 'xlabel':'Time(s)', 'ylabel':'Amplitude(m)'}]

getReadySubplot(title='Multiplication of Two Signal', signals=signals, figSize=(15,18))
plt.show()
```

<p style="font-size:20px;">$$\begin{aligned}&(iv) Amplitude\ and\ Time\ Scaling\end{aligned} $$</p>

```python
def originalSignal(times, omega):
    return [sin(omega * t) for t in times]

def amplitudeScaling(times, omega, scalingFactor):
    return [scalingFactor * sin(omega * t) for t in times]

def timeScaling(times, omega, scalingFactor):
    return [sin(omega * t * scalingFactor) for t in times]
```

```python
T = 4
omega = (2 * pi)/T
times = np.arange(0,3 * T, 0.01)

signals = [{'title':'$sin(t)$','x':times, 'y':originalSignal(times,omega), 'xlabel':'Time(s)', 'ylabel':'Amplitude(m)'}, 
           {'title':'$2sin(t)$','x':times, 'y':amplitudeScaling(times,omega,2), 'xlabel':'Time(s)', 'ylabel':'Amplitude(m)'},
           {'title':'$sin(0.5t)$','x':times, 'y':timeScaling(times,omega,0.5), 'xlabel':'Time(s)', 'ylabel':'Amplitude(m)'}]

getReadySubplot(title='Amplitude and Time Scaling of a Signal', signals=signals, figSize=(15,18))
plt.show()
```

<p style="font-size:20px;">$$\begin{aligned}&(v) Odd\ and\ Even\ Components\ of\ a\ Signal\end{aligned} $$</p>

```python
def originalSignal(times, omega):
    return [sin(t * omega) + cos(3 * t * omega) for t in times]

def oddComponent(orgSgnl, revOrgSgnl):
    return [0.5 * (i - j) for i,j in zip(orgSgnl,revOrgSgnl)]

def evenComponent(orgSgnl, revOrgSgnl):
    return [0.5 * (i + j) for i,j in zip(orgSgnl,revOrgSgnl)]

def recoverOriginal(even, odd):
    return [e + o for e,o in zip(even,odd)]
    
```

```python
T = 4
omega = (2 * pi)/T
times = np.arange(0,3 * T, 0.01)

orgSgnl = originalSignal(times, omega)
revOrgSgnl = originalSignal(-times, omega)

even = evenComponent(orgSgnl,revOrgSgnl)
odd = oddComponent(orgSgnl,revOrgSgnl)

signals = [{'title':'$Original\ Signal$','x':times, 'y':originalSignal(times,omega), 'xlabel':'Time(s)', 'ylabel':'Amplitude(m)'}, 
           {'title':'$Even\ Component$','x':times, 'y':even, 'xlabel':'Time(s)', 'ylabel':'Amplitude(m)'},
           {'title':'$Odd\ Component$','x':times, 'y':odd, 'xlabel':'Time(s)', 'ylabel':'Amplitude(m)'},
           {'title':'$Recovered\ Original\ Signal$','x':times, 'y':recoverOriginal(even,odd), 'xlabel':'Time(s)', 'ylabel':'Amplitude(m)'}]

getReadySubplot(title='Even Odd Components of a Signal', signals=signals, figSize=(15,20))
plt.show()
```
