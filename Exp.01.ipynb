<p style="font-size:20px">Import libraries and plot setup.</p>

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from math import pi,cos,e
```

```python
def getPlotReady(title='',xLabel='', yLabel='', figSize=None):
    fig = plt.figure()
    if figSize != None:
        fig.set_size_inches(figSize[0], figSize[1])
        
    ax = fig.add_subplot(1,1,1)
    ax.set_title(title,fontdict={'fontsize': 20}, pad=15.0)
    ax.set_xlabel(xLabel,fontdict={'fontsize': 16})
    ax.set_ylabel(yLabel,fontdict={'fontsize': 16})
    
```

```python
def getPlotReadyForMultipleSignal(signals,title='',figSize=None):
    fig, ax = plt.subplots()
    if figSize != None:
        fig.set_size_inches(figSize[0], figSize[1])
    for signal in signals:
        ax.plot(signal['x'], signal['y'], 
            color=signal['color'], 
            linewidth=signal['linewidth'],
            label=signal['name'])

# Enable legend
    ax.legend()
    ax.set_title(title,fontdict={'fontsize': 20}, pad=15.0)
    ax.set_xlabel("Time(t)",fontdict={'fontsize': 16})
    ax.set_ylabel("Amplitude(m)",fontdict={'fontsize': 16})
    
```

```python
def getReadySubplot(numOfPlot, signals, figSize=None):
    fig, axes = plt.subplots(numOfPlot, 1) 
    if figSize != None:
        fig.set_size_inches(figSize[0], figSize[1])
    fig.suptitle('Multiple Signal in subplot.', fontsize=20, y=0.93)
    for index in range(numOfPlot):
        axes[index].plot(signals[index]['x'],signals[index]['y'])
        axes[index].set_title(signals[index]['title'])
        axes[index].set_xlabel(signals[index]['xlabel'])
        axes[index].set_ylabel(signals[index]['ylabel'])
```

<p style="font-size:20px;">$$\begin{aligned}&(i) y(t) = cos(t), 0<=t<=10 \end{aligned} $$</p>

```python
def get_amp_1(times):
    return [cos(t) for t in times]
```

```python
times = np.arange(0,10,0.01)
amp = get_amp_1(times)
getPlotReady(title='cos(t)', xLabel='Times(s)',yLabel='Amplitude', figSize=(15,8))
plt.plot(times,amp)
```

<p style="font-size:20px;">$$\begin{aligned}&(ii) y(t) = 3cos(3\pi t + \pi/3), in four periods \end{aligned} $$</p>

```python
def get_amp_2(times,A,omega,phi):
    return [A*cos(omega * t + phi) for t in times]
```

```python
A = 3
phi = pi/3
omega = 3 * pi
T = (2 * pi)/omega
times = np.arange(0,4*T,0.01)
amp = get_amp_2(times,A,omega,phi)
getPlotReady(title='$3cos(3\pi t + \pi/3)$', xLabel='Times(s)',yLabel='Amplitude', figSize=(15,8))
plt.plot(times,amp)
```

<p style="font-size:20px;">$$\begin{aligned}&(iii) y(t) = 2e^{-0.9t}, 0<=t<=10 \end{aligned} $$</p>

```python
def get_amp_3(times):
    return [2 * e ** (-0.9 * t) for t in times]
```

```python
times = np.arange(0,10,0.01)
amp = get_amp_3(times)
getPlotReady(title='$2e^{-0.9t}$', xLabel='Times(s)',yLabel='Amplitude', figSize=(15,8))
plt.plot(times,amp)
```

<p style="font-size:20px;">Ploting all the signal in a single figure. </p>

```python
signals = [{'title': '$cos(t)$', 'x': np.arange(0,10,0.01),
             'y': get_amp_1(np.arange(0,10,0.01)), 'xlabel':'Time(s)', 'ylabel':'Amplitude(m)'},
            {'title': '$3cos(3\pi t + \pi/3)$', 'x': np.arange(0,4*T,0.01),
             'y': get_amp_2(np.arange(0,4*T,0.01),A,omega,phi), 'xlabel':'Time(s)', 'ylabel':'Amplitude(m)'},
            {'title': '$2e^{-0.9t}$', 'x': np.arange(0,10,0.01),
             'y': get_amp_3(np.arange(0,10,0.01)), 'xlabel':'Time(s)', 'ylabel':'Amplitude(m)'}]

getReadySubplot(numOfPlot=3, signals=signals, figSize=(15,20))
plt.show()
```
