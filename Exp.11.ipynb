x(t)\ =\ cos(2\pi * 47t)\ +\ cos(2\pi * 219t)\ when\ 0 <= t <= 10\ otherwise\ 0.$$</i></p>

<p style="font-size:20px">Import libraries and plot setup.</p>

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from numpy import sin, cos, pi, exp
from scipy.signal import welch
from scipy.integrate import simps as intgrt
```

```python
def getPlotReady(title='',xLabel='', yLabel='', figSize=None):
    fig = plt.figure()
    if figSize != None:
        fig.set_size_inches(figSize[0], figSize[1])
        
    ax = fig.add_subplot(1,1,1)
    ax.set_title(title,fontdict={'fontsize': 20}, pad=15.0)
    ax.set_xlabel(xLabel,fontdict={'fontsize': 16})
    ax.set_ylabel(yLabel,fontdict={'fontsize': 16})
    
```

<p style="font-size:20px">Generate the signal.</p>

```python
samples = 1000

omega1 = (2.0 * pi * 47)
omega2 = (2.0 * pi * 219)

T1 = (2 * pi)/omega1
print(T1)
T2 = (2 * pi)/omega2
print(T2)

T = int(max(T1, T2))
```

```python
T
```

```python
# t = np.linspace(0, T, samples * (T-0), endpoint=False)
t = np.linspace(0, 10, samples * 10, endpoint=False)
signal = cos(omega1 * t) + cos(omega2 * t)
```

```python
freqs, psd = welch(signal)

getPlotReady(title='$cos(2\pi * 47t)\ +\ cos(2\pi * 219t)$', xLabel='Frequency',yLabel='Power', figSize=(15,8))
plt.semilogx(freqs, psd)
# plt.plot(freqs, psd)
# print(freqs)
# plt.plot(t,signal)
```
